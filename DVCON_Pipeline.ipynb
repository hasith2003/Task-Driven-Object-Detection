```python
import sys
import os

# 1. Add the 'src' directory to the path so Python can find the modules
# Replace this with the actual path to your 'src' folder
src_path = '/content/drive/MyDrive/task-driven-object-detection/src'
if src_path not in sys.path:
    sys.path.append(src_path)

# 2. Now import the module
import PIL.Image as Image
import coco_tasks.single_task_datasets

# 3. Define the new, "safe" function
def safe_load_image(file_name):
    if not os.path.exists(file_name):
        print(f"⚠️ Missing file skipped: {file_name}")
        return Image.new('RGB', (224, 224), (0, 0, 0))
    try:
        return Image.open(file_name).convert('RGB')
    except Exception as e:
        print(f"❌ Error loading {file_name}: {e}")
        return Image.new('RGB', (224, 224), (0, 0, 0))

# 4. Overwrite it
coco_tasks.single_task_datasets.load_image = safe_load_image

print("✅ Patch applied successfully, bros!")
```
